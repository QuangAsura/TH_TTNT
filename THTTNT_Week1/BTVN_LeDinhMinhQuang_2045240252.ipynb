```python
import numpy as np
#Bai Tap Ve Nha
# Bai 1
a1 = [3, 9, 1, 4]
b1 = [2, 7, 4, 3, 2, 8]
mang_kq = []
min_len = min(len(a1), len(b1))
for i in range(min_len):
    mang_kq.append(min(a1[i], b1[i]))
if len(a1) > len(b1):
    mang_kq.extend(a1[min_len:])
else:
    mang_kq.extend(b1[min_len:])
print("Mang ket qua sau khi tron: ", mang_kq)
n = 4
manganew = np.random.randint(-10, 20, (n, n))
print("Ma tran ban dau:\n", manganew)
```

```python
import numpy as np
#bai2
# a. Tong cac phan tu thuoc tam giac tren cua duong cheo phu (ke ca duong cheo phu)
tong_tg_tren = 0
for i in range(n):
    for j in range(n):
        if i + j <= n - 1:
            tong_tg_tren += manganew[i][j]
print("\nTong tam giac tren duong cheo phu: ", tong_tg_tren)

# b. Chuyen phan tu am thanh tri tuyet doi
manganew = np.abs(manganew)
print("\nMang sau khi chuyen tri tuyet doi:\n", manganew)

# c. Thay phan tu chan bang so x cho truoc
x = 99
for i in range(n):
    for j in range(n):
        if manganew[i][j] % 2 == 0:
            manganew[i][j] = x
print(f"\nMang sau khi thay so chan bang {x}:\n", manganew)

# d. Kiem tra a co toan chan khong?
check_toan_chan = np.all(manganew % 2 == 0)
print("\nMang co toan chan khong? ", check_toan_chan)

# e. Kiem tra ma tran doi xung (a[i][j] == a[j][i])
is_doi_xung = True
for i in range(n):
    for j in range(n):
        if manganew[i][j] != manganew[j][i]:
            is_doi_xung = False
            break
print("\nMa tran co doi xung khong? ", is_doi_xung)

# f. Kiem tra duong cheo chinh tang dan
tang_dan = True
for i in range(n - 1):
    if manganew[i][i] >= manganew[i+1][i+1]:
        tang_dan = False
        break
print("\nDuong cheo chinh co tang dan khong? ", tang_dan)

# g. Xuat phan tu thuoc tam giac duoi cua duong cheo phu
print("\nCac phan tu tam giac duoi duong cheo phu:")
for i in range(n):
    for j in range(n):
        if i + j >= n - 1:
            print(manganew[i][j], end=" ")

# h. Kiem tra duong cheo phu giam dan
giam_dan = True
for i in range(n - 1):
    if manganew[i][n-1-i] <= manganew[i+1][n-2-i]:
        giam_dan = False
        break
print("\n\nDuong cheo phu co giam dan khong? ", giam_dan)
```
