```python
# Bai 3 Xóa khoảng trắng thừa ở đầu và cuối chuỗi “ Hello World ”. Chuyển chuỗi sang dạng hoa và thường. Thay ký tự H thành ký tự J.
original_string = " Hello World "
print(f"Original string: '{original_string}'")

stripped_string = original_string.strip()
print(f"String after stripping whitespace: '{stripped_string}'")

uppercase_string = stripped_string.upper()
print(f"String in uppercase: '{uppercase_string}'")

lowercase_string = stripped_string.lower()
print(f"String in lowercase: '{lowercase_string}'")

replaced_string = stripped_string.replace('H', 'J')
print(f"String after replacing 'H' with 'J': '{replaced_string}'")
```

```python
#Bai 14
import numpy as np
m = int(input("Nhap vao m dong: "))
n = int(input("Nhap vao n cot: "))
#Cau a
manganew = np.random.rand(m,n)
print("mang a sau khi random: ",manganew)

#Cau b
k = int(input("nhap vao dong k ma ban muon: "))
if 0 <= k < m:
    print("\nCac phan tu thuoc dong k: ", manganew[k])

#Cau c:
k1 = int(input("nhap vao cot k ma ban muon lay ra: "))
if 0 <= k1 < n:
    k1_col = [row[k1] for row in manganew]
    print("\n cac phan tu thuoc cot k la: ", k1_col)

#Cau d:
row_sums = np.sum(manganew, axis=1)
max_row_index = np.argmax(row_sums)
print("\nDong co tong lon nhat la:", manganew[max_row_index])
print("Tong cua dong do la:", row_sums[max_row_index])

#cau e:
col_products = np.prod(manganew, axis=0)
min_col_index = np.argmin(col_products)
print("\nCot co tich nho nhat la:", manganew[:, min_col_index])
print("Tich cua cot do la:", col_products[min_col_index])

#cau f:
print("\nCac phan tu thuoc dong chan va cot le:")
for i in range(m):
    for j in range(n):
        if i % 2 == 0 and j % 2 != 0:
            print(f"a[{i}][{j}] = {manganew[i][j]}")

#cau g:
mang_chan_dong_le = []
for i in range(m):
    if i % 2 != 0:
        for j in range(n):
            if int(manganew[i][j]) % 2 == 0:
                mang_chan_dong_le.append(manganew[i][j])
if len(mang_chan_dong_le) > 0:
    print("\nTrung binh cong:", sum(mang_chan_dong_le)/len(mang_chan_dong_le))

#Cau h:
mang_bien = []
for i in range(m):
    for j in range(n):
        if i == 0 or i == m-1 or j == 0 or j == n-1:
            mang_bien.append(manganew[i][j])
print("\nTrung binh cong bien:", sum(mang_bien)/len(mang_bien))

#cau i:
mang_trong = []
for i in range(1, m-1):
    for j in range(1, n-1):
        mang_trong.append(manganew[i][j])
if len(mang_trong) > 0:
    tich = 1
    for x in mang_trong:
        tich *= x
    print("\nTrung binh tich:", tich ** (1/len(mang_trong)))
```

```python
#Bai 15
class SinhVien:
    def __init__(self, n):
        self.dic_SinhVien = {'maSV': [], 'hoTen': [], 'namsinh': [], 'diemTB': []}
        for i in range(n):
            self.dic_SinhVien['maSV'].append(input("Nhap ma SV: "))
            self.dic_SinhVien['hoTen'].append(input("Nhap ho ten: "))
            self.dic_SinhVien['namsinh'].append(int(input("Nhap nam sinh: ")))
            self.dic_SinhVien['diemTB'].append(float(input("Nhap diem: ")))

sv_manager = SinhVien(2)
print(sv_manager.dic_SinhVien)

#c
def check_pass(dic):
    return sum(d >= 5 for d in dic['diemTB'])
print(check_pass(sv_manager.dic_SinhVien))

#d
def check_age(dic):
    for i in range(len(dic['namsinh'])):
        if 2026 - dic['namsinh'][i] >= 20:
            print(dic['hoTen'][i])
check_age(sv_manager.dic_SinhVien)

#e
def count_dh_students(dic):
    return sum(len(ma)>=4 and ma[2:4].upper()=="DH" for ma in dic['maSV'])
print(count_dh_students(sv_manager.dic_SinhVien))

#f
def count_name_lan(dic):
    return sum(name.split()[-1].lower()=="lan" for name in dic['hoTen'])
print(count_name_lan(sv_manager.dic_SinhVien))

#g
def count_surname_phan(dic):
    return sum(name.split()[0].lower()=="phan" for name in dic['hoTen'])
print(count_surname_phan(sv_manager.dic_SinhVien))
```
