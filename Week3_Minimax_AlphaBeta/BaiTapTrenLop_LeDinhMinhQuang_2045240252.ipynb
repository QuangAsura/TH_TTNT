```python
import copy
import math
import numpy

X = "X"
O = "O"
EMPTY = None

def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    # Đếm số ô đã đánh. X luôn đi trước.
    count = 0
    for row in board:
        for cell in row:
            if cell is not EMPTY:
                count += 1
    # Nếu số nước cờ trên bàn là số chẵn (0, 2, 4...) thì đến lượt X
    if count % 2 == 0:
        return X
    return O

def actions(board):
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len): # Đã thêm vòng lặp j bị thiếu
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def get_horizontal_winner(board):
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
                break
        if winner_val is not None:
            return winner_val
    return None

def get_vertical_winner(board):
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
                break
        if winner_val is not None:
            return winner_val
    return None

def get_diagonal_winner(board):
    board_len = len(board)

    # Check đường chéo chính
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
            break
    if winner_val is not None:
        return winner_val

    # Check đường chéo phụ
    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None
            break
    return winner_val

def winner(board):
    return get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board)

def terminal(board):
    if winner(board) is not None:
        return True
    for row in board:
        for cell in row:
            if cell == EMPTY:
                return False
    return True

def utility(board):
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    current_player = player(board)
    move = None
    if current_player == X:
        best_val = -math.inf
        for action in actions(board):
            check = minValue(result(board, action))
            if check > best_val:
                best_val = check
                move = action
    else:
        best_val = math.inf
        for action in actions(board):
            check = maxValue(result(board, action))
            if check < best_val:
                best_val = check
                move = action
    return move

if __name__ == "__main__":
    board = initial_state()

    print("Choose a player (X or O):")
    user = input().strip().upper()
    while user not in [X, O]:
        print("Invalid! Please choose X or O:")
        user = input().strip().upper()

    if user == X:
        ai = O
    else:
        ai = X

    print(f"You play as {user}. AI plays as {ai}.")
    print("X always goes first!\n")

    while True:
        game_over = terminal(board)
        current_turn = player(board)

        if game_over:
            win_player = winner(board)
            if win_player is None:
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {win_player} wins.")
            break

        if current_turn == ai:
            print("AI is thinking...")
            move = minimax(board)
            if move:
                board = result(board, move)
            print(numpy.array(board))
            print("-" * 15)
        else:
            print(f"Your turn ({user}). Enter position to move (0-2):")
            try:
                i = int(input("Row: "))
                j = int(input("Col: "))
                if i in [0, 1, 2] and j in [0, 1, 2] and board[i][j] == EMPTY:
                    board = result(board, (i, j))
                    print(numpy.array(board))
                    print("-" * 15)
                else:
                    print("Invalid move. Cell is occupied or out of bounds. Try again.")
            except ValueError:
                print("Please enter a valid integer.")
```

```python
import os
import math

def GetWinner(board):
    # Các hàng ngang
    for i in range(0, 9, 3):
        if board[i] == board[i+1] == board[i+2]: return board[i]
    # Các cột dọc
    for i in range(3):
        if board[i] == board[i+3] == board[i+6]: return board[i]
    # Các đường chéo
    if board[0] == board[4] == board[8]: return board[0]
    if board[2] == board[4] == board[6]: return board[2]
    return None

def PrintBoard(board):
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f'''
     {board[0]} | {board[1]} | {board[2]}
    -----------
     {board[3]} | {board[4]} | {board[5]}
    -----------
     {board[6]} | {board[7]} | {board[8]}
    ''')

def GetAvailableCells(board):
    # Chỉ lấy những ô vẫn còn là số (chưa bị đánh X hoặc O)
    return [cell for cell in board if isinstance(cell, int)]

def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner == "X": return 10 - depth
    if winner == "O": return -10 + depth

    available = GetAvailableCells(position)
    if len(available) == 0: return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in available:
            position[cell - 1] = "X"
            eval = minimax(position, depth + 1, alpha, beta, False)
            position[cell - 1] = cell
            maxEval = max(maxEval, eval)
            alpha = max(alpha, eval) # Cập nhật Alpha
            if beta <= alpha: break
        return maxEval
    else:
        minEval = math.inf
        for cell in available:
            position[cell - 1] = "O"
            eval = minimax(position, depth + 1, alpha, beta, True)
            position[cell - 1] = cell
            minEval = min(minEval, eval)
            beta = min(beta, eval) # Cập nhật Beta
            if beta <= alpha: break
        return minEval

def FindBestMove(currentPosition, AI):
    bestVal = -math.inf if AI == "X" else math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        # Nếu AI là X thì lượt tiếp theo trong minimax phải là Maximizing = False
        moveVal = minimax(currentPosition, 0, -math.inf, math.inf, AI == "O")
        currentPosition[cell - 1] = cell

        if AI == "X":
            if moveVal > bestVal:
                bestVal = moveVal
                bestMove = cell
        else:
            if moveVal < bestVal:
                bestVal = moveVal
                bestMove = cell
    return bestMove

def main():
    player = input("Play as X or O? ").strip().upper()
    if player not in ["X", "O"]: player = "X"

    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)]
    currentTurn = "X" # X luôn đi trước theo luật quốc tế

    while True:
        winner = GetWinner(currentGame)
        available = GetAvailableCells(currentGame)

        if winner or not available:
            PrintBoard(currentGame)
            if winner: print(f"{winner} THẮNG!!!")
            else: print("HÒA!")
            break

        if currentTurn == AI:
            print("AI đang suy nghĩ...")
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        else:
            PrintBoard(currentGame)
            try:
                move = int(input(f"Lượt của bạn ({player}). Nhập số ô: "))
                if move in currentGame:
                    currentGame[move - 1] = player
                    currentTurn = AI
                else:
                    print("Ô không hợp lệ!")
                    continue
            except ValueError:
                print("Vui lòng nhập số!")
                continue

if __name__ == "__main__":
    main()
```
